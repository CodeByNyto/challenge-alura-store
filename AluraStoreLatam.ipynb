#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda1['Precio'].sum()
print(f"Ingreso total de la tienda 1: {ingreso_total_tienda1}")

ingreso_total_tienda2 = tienda2['Precio'].sum()
print(f"Ingreso total de la tienda 2: {ingreso_total_tienda2}")

ingreso_total_tienda3 = tienda3['Precio'].sum()
print(f"Ingreso total de la tienda 3: {ingreso_total_tienda3}")

ingreso_total_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso total de la tienda 4: {ingreso_total_tienda4}")
```

```python
from numpy._core.defchararray import center
tiendas = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
ingresos = [ingreso_total_tienda1,ingreso_total_tienda2,ingreso_total_tienda3,ingreso_total_tienda4]

plt.title('Ingreso Total por Tienda',
          loc='center',
          fontsize=12,
          fontweight='bold',
          pad=14
)

plt.bar(tiendas, ingresos, color=['lightpink'])
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales en Millones')

plt.savefig('grafico1_ingresos_totales_tiendas.png', dpi=300, bbox_inches='tight')
```

Se identifica que la Tienda 4 obtuvo menos ingresos totales en comparación a las otras 3 tiendas.

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por Categoría de la Tienda 1:")
print(ventas_por_categoria_tienda1)

ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nVentas por Categoría de la Tienda 2:")
print(ventas_por_categoria_tienda2)

ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nVentas por Categoría de la Tienda 3:")
print(ventas_por_categoria_tienda3)

ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nVentas por Categoría de la Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = (tienda1['Calificación'].mean())
print(f"La calificación promedio de la Tienda 1 es: {calificacion_promedio_tienda1:.2f}")

calificacion_promedio_tienda2 = (tienda2['Calificación'].mean())
print(f"La calificación promedio de la Tienda 2 es: {calificacion_promedio_tienda2:.2f}")

calificacion_promedio_tienda3 = (tienda3['Calificación'].mean())
print(f"La calificación promedio de la Tienda 3 es: {calificacion_promedio_tienda3:.2f}")

calificacion_promedio_tienda4 = (tienda4['Calificación'].mean())
print(f"La calificación promedio de la Tienda 4 es: {calificacion_promedio_tienda4:.2f}")
```

```python
calificaciones_promedio = {'Tienda 1': calificacion_promedio_tienda1, 'Tienda 2': calificacion_promedio_tienda2, 'Tienda 3': calificacion_promedio_tienda3, 'Tienda 4': calificacion_promedio_tienda4}
df_calificaciones_promedio = pd.DataFrame(list(calificaciones_promedio.items()), columns=['Tienda', 'Promedio de Calificacion'])

plt.figure(figsize=(10,6))
sns.pointplot(x='Tienda', y='Promedio de Calificacion', data=df_calificaciones_promedio)
plt.title('Calificación Promedio por Tienda', loc='center', fontsize=12, fontweight='bold',pad=14)
plt.xlabel('Tienda')
plt.ylabel('Promedio de Calificacion')

plt.savefig('grafico2_calificacion_promedio_tiendas.png', dpi=300, bbox_inches='tight')
```

Se evidencia que la Tienda 1 presenta un promedio de 3.98, calificación más baja entre las 4 tiendas

# 4. Productos más y menos vendidos

```python
conteo_productos = tienda1['Producto'].value_counts()
print(f"Top 3 productos más vendidos en la Tienda 1:")
print(conteo_productos.head(3))

print(f"\nTop 3 productos menos vendidos en la Tienda 1:")
print(conteo_productos.tail(3))

conteo_productos = tienda2['Producto'].value_counts()
print(f"\nTop 3 productos más vendidos en la Tienda 2:")
print(conteo_productos.head(3))

print(f"\nTop 3 productos menos vendidos en la Tienda 2:")
print(conteo_productos.tail(3))

conteo_productos = tienda3['Producto'].value_counts()
print(f"\nTop 3 productos más vendidos en la Tienda 3:")
print(conteo_productos.head(3))

print(f"\nTop 3 productos menos vendidos en la Tienda 3:")
print(conteo_productos.tail(3))

conteo_productos = tienda4['Producto'].value_counts()
print(f"\nTop 3 productos más vendidos en la Tienda 4:")
print(conteo_productos.head(3))

print(f"\nTop 3 productos menos vendidos en la Tienda 4:")
print(conteo_productos.tail(3))
```

```python
dataframes_tiendas = [tienda1, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1','Tienda 2','Tienda 3','Tienda 4']

def productos_mas_menos_vendidos(df,tiendas, n=3):

  conteo_productos = df['Producto'].value_counts()
  mas_n = conteo_productos.head(n)
  menos_n = conteo_productos.tail(n)

  productos = pd.concat([mas_n, menos_n]).sort_values(ascending=False)

  colores = ['lightpink'] * len(mas_n) + ['lightblue'] * len(menos_n)

  if len(productos) < (2 * n):
    colores = ['lightpink'] * len(mas_n) + ['lightblue'] * (len(productos) - len(mas_n))

  plt.figure(figsize=(12, 6))

  bars = plt.bar(productos.index, productos.values, color=colores)

  for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{int(yval)}', ha='center', va='bottom', fontsize=9)

  plt.title(f'Productos más y menos vendidos por Tienda', fontsize=15)
  plt.xlabel('Producto', fontsize=13)
  plt.ylabel('Cantidad de Ventas', fontsize=13)
  plt.show()

for df, nombre in zip(dataframes_tiendas, tiendas):
  productos_mas_menos_vendidos(df, nombre)

plt.savefig('grafico3_productos_mas_menos_vendidos_tiendas.png', dpi=300, bbox_inches='tight')
```
